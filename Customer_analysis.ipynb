```python
df.drop(columns=['ID','Dt_Customer'],inplace =True)
df
```

```python
df.info()
```

Checking the null values

```python
df.isna().sum()
```

```python
missing=df.groupby(['Education','Marital_Status'])['Income'].transform("mean").round(0)
df["Income"].fillna(missing, inplace=True)
```

```python
df.describe().round(2)
```

Check if there's any duplication

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
average_income = df['Income'].mean()
average_income
```

```python
income_by_kids = df.groupby('Kidhome')['Income'].mean()
plt.figure(figsize=(8,6))
income_by_kids.plot(kind='bar')
plt.title('Average Income by Number of kids at home')
plt.xlabel('Number of Kids at home')
plt.ylabel('Average Income')
plt.show()
```

The above graph highlights that the average income of the customers is also affected by the number of kids at home. Those who have zero kids at home are the one with highest average income

```python
income_by_maritalstatus=df.groupby('Marital_Status')['Income'].mean()
income_by_maritalstatus
```

```python
plt.figure(figsize=(8,6))
income_by_maritalstatus.plot(kind='bar',color='pink')
plt.title('Average Income by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Average Income')
plt.show()
```
